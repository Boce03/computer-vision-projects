```python
import numpy as np
```

```python
def aff2hom(pt):
    return np.append(pt, 1)

def hom2aff(pt):
    return pt[:-1] / pt[2]
```

```python
# afine(piksel) koordinate
p1 = np.array([812,421])
p2 = np.array([474, 590])
p3 = np.array([380, 488])
p5 = np.array([848, 235])
p6 = np.array([458, 365])
p7 = np.array([354, 285])
p8 = np.array([715, 190])

p1h = aff2hom(p1)
p2h = aff2hom(p2)
p3h = aff2hom(p3)
p5h = aff2hom(p5)
p6h = aff2hom(p6)
p7h = aff2hom(p7)
p8h = aff2hom(p8)
```

```python
#osmoteme([p1,p2,p3,p5,p6,p7,p8])
def osmoteme(pt):
    p67 = np.cross(pt[4], pt[5])
    p23 = np.cross(pt[1], pt[2])
    p = np.cross(p23, p67)
    p1 = np.cross(pt[0], p)

    q15 = np.cross(pt[0], pt[3])
    q26 = np.cross(pt[1], pt[4])
    q = np.cross(q15, q26)
    q8 = np.cross(q, pt[6])

    p4h = np.cross(q8,p1)
    return aff2hom(np.round(hom2aff(p4h)))
```
